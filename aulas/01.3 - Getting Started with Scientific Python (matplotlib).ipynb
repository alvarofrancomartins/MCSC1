#### Adding a legend

````python
plt.plot(x,np.cos(x),label=r'$f(x)=cos(x)$')
plt.plot(x,np.sin(x),label=r'$f(x)=sin(x)$')
plt.legend()
#plt.legend(loc='upper left')
````

 #### Annotate some points

````python
plt.plot(x,np.cos(x),label=r'$f(x)=cos(x)$')
plt.plot(x,np.sin(x),label=r'$f(x)=sin(x)$')
plt.legend()
t=np.pi/2
plt.annotate(r'$sin\left(\frac{\pi}{2}\right)=1$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(60, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=1"))
````

#### Other types of plots

````python
import numpy as np
from matplotlib import pyplot as plt
````

````python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)
Y2 = np.sin(3 * X)
````

#### Plot and filled plots

````python
f, ax = plt.subplots(figsize=(5,4))
ax.plot(X,Y+1,color='red')
ax.fill_between(X,Y+1,Y+1.5,color='red',alpha=0.5,linewidth=0)
ax.fill_between(X,Y+1,Y+0.5,color='red',alpha=0.5,linewidth=0)

ax.plot(X,Y2-0.5,color='blue')
ax.fill_between(X,Y2-0.5,Y2+0,color='blue',alpha=0.5,linewidth=0)
ax.fill_between(X,Y2-0.5,Y2-1,color='blue',alpha=0.5,linewidth=0)
````

#### Scatter Plots

````python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
f, ax = plt.subplots(figsize=(5,4))
ax.scatter(X,Y)
````

````python
dis = np.sqrt(X**2+Y**2)
````

````python
f, ax = plt.subplots(figsize=(5,4))
ax.scatter(X,Y,c=dis,cmap='magma_r')
````

````python
f, ax = plt.subplots(figsize=(5,4))
ax.scatter(X,Y,c=dis,cmap='magma_r',s=40*dis)
````

#### Bar Plots

````python
data = [10,4,16,22]
names = ['aaa','bbb','ccc','ddd']
f, ax = plt.subplots(figsize=(5,4))
plt.xticks(())
ax.bar(np.arange(len(data)),data)
for x,y,name in zip(np.arange(len(data)),data,names):
    ax.text(x,y+0.2,'{}'.format(name), ha='center')
````

#### Contour Plots

##### `np.meshgrid`

> creates a rectangular grid out of an array of x values and an array of y values.

````python
x = np.array([0, 1, 2, 3, 4]);
y = np.array([0, 1, 2, 3, 4, 5]);
X, Y = np.meshgrid(x, y)
plt.plot(X,Y,marker='>',linestyle='none')
````

````python
def gauss2d(x,y):
    return(np.exp(-x**2-y**2))
````

````python
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)
````

````python
f, ax = plt.subplots(figsize=(5,4))
ax.contourf(X, Y, gauss2d(X, Y), 20,
            alpha=.75, cmap='viridis')
````

````python
f, ax = plt.subplots(figsize=(5,4))
ct = ax.contour(X, Y, gauss2d(X, Y), 4,
           alpha=.75,colors='k')
ct.clabel(fontsize=8)
````

````python
f, ax = plt.subplots(figsize=(5,4))

ax.contourf(X, Y, gauss2d(X, Y), 4,
            alpha=1, cmap='viridis')
ct = ax.contour(X, Y, gauss2d(X, Y), 4,
           alpha=.75,colors='k')
ct.clabel(fontsize=8)
````

#### Imshow

````python
f, ax = plt.subplots(figsize=(5,4))
ax.imshow(gauss2d(X, Y),cmap='Blues_r')
````
